# Histogram
- To draw no of observations in a given interval.
- It is used to represent the frequency distribution.
- hist() is used to create Histograms.
- it is very helpful to analyze data with simple steps.

```python
# import required libraries
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Simple example of Histograms
plt.figure(figsize=(10,8))
x = np.random.rand(10000)
plt.hist(x)
plt.show()
```

```python
# Custom X-ticks and bins of histogram and how to set edge color.
plt.figure(figsize=(10,8))
data = [np.random.randint(1,7)+np.random.randint(1,7)+np.random.randint(1,7) for i in range(100000)]
bins = np.arange(1,19) #[1,2,3,...,18]
plt.hist(data,bins=bins,ec='red')
plt.xticks(bins)
plt.grid(axis='y')
plt.show()
```

**How to set Color of Bins in Histograms**
- hist() function returns a tuple with lenth 3.
- n -> no of observations available in each interval.
- bins -> no of bins.
- patches -> object of BarContainer.

```python
# set the color of first bin
plt.figure(figsize=(10,8))
data = [np.random.randint(1,7)+np.random.randint(1,7)+np.random.randint(1,7) for i in range(100000)]
bins = np.arange(1,19) #[1,2,3,...,18]
hist_returns = plt.hist(data,bins=bins,ec='red')
hist_returns[2][2].set_fc('yellow')
plt.xticks(bins)
plt.grid(axis='y')
plt.show()
```

```python
# Set color for ach bin
colors = ['r','g','orange','b','k','m','y','c','r','g','orange','b','k','m','y','c','g','b']
plt.figure(figsize=(10,8))
data = [np.random.randint(1,7)+np.random.randint(1,7)+np.random.randint(1,7) for i in range(100000)]
bins = np.arange(1,19) #[1,2,3,...,18]
hist_returns = plt.hist(data,bins=bins,ec='red')
for bin in range(hist_returns[0].size):
    hist_returns[2][bin].set_fc(colors[bin])
plt.xticks(bins)
plt.grid(axis='y')
plt.show()
```

# Scatter plots
- The most commonly used plot of matplotlib.
- Very simple to understand.
- Very helpful to know the relationship between two variables.
- scatter() function is used to plot.
- The terminology is very much similar to line plot that we have seen in Part-1 of this tutorial series.


```python
#  A simple example of scatter plot.
plt.figure(num=1,figsize=(18,8),facecolor="green")
x = np.arange(20)
plt.scatter(x,x**2)
plt.show()
```

**Colors and Size of Scatter Plot**

```python
# Increase the size and color for each point.
plt.figure(num=1,figsize=(18,8),facecolor="green")
x = np.arange(5)
colors = ['r','g','b','y','orange']
size = np.random.randint(300,700,5)
plt.scatter(x,x**2,c=colors,s=size,)
plt.show()
```

**ColorMap of Scatterplot**
- There are many colormaps you can check it out https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
x = np.random.randint(1,101,size=(100))
y = np.random.randint(1,101,size=(100))
sizes = 5*np.random.randint(1,101,size=(100))
colors = np.random.randint(1,100,100)
plt.scatter(x,y,s=sizes,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
# https://matplotlib.org/stable/tutorials/colors/colormaps.html
x = np.random.randint(1,101,size=(100))
y = np.random.randint(1,101,size=(100))
sizes = 5*np.random.randint(1,101,size=(100))
colors = np.random.randint(1,100,100)
plt.scatter(x,y,s=sizes,c=colors,cmap='flag')
plt.colorbar()
plt.show()
```

# Text and Annotate in scatter Plot

```python
# Increase the size and color for each point.
plt.figure(num=1,figsize=(18,8),facecolor="green")
x = np.arange(5)
labels = ['Label-'+str(i) for i in range(1,6)]
colors = ['r','g','b','y','orange']
size = np.random.randint(300,700,5)
scatter = plt.scatter(x,x**2,c=colors,s=size,cmap='flag')
plt.legend(handles = scatter.legend_elements()[0],labels=labels)
plt.colorbar()
plt.show()
```

```python
plt.figure(num=1,figsize=(18,8),facecolor="green")
x = np.arange(5)
y = x
labels = ['Label-'+str(i) for i in range(1,6)]
colors = ['r','g','b','y','orange']
size = np.random.randint(300,700,5)
plt.scatter(x,y,c=colors,s=size,cmap='flag')
for i,label in enumerate(labels):
    plt.text(x[i]+0.2,y[i],labels[i],ha='center')
plt.show()
```

# SubPlots

**Two Important Terms**
- Figure --> it is a window in which matplotlib shows the graps and plots.
- Axes --> it is the area under figure where we can set x-axis,y-axis,bar,plot,hist etc.

```python
figure = plt.figure(figsize=(10,6),facecolor='blue')
ax = figure.add_axes([0.1,0.1,0.8,0.8]) #[l,b,w,h]
plt.show()
```

**The colored portion is figure and the white called axes**
- we can set axes,plots,title etc for axes.
- see in example below

```python
figure = plt.figure(figsize=(10,6),facecolor='blue')
x = np.arange(6)
y = x**2
ax = figure.add_axes([0.1,0.1,0.8,0.8]) #[l,b,w,h]
ax.plot(x,y,'o--g',markersize=20)
ax.set_title("x vs y",fontdict={'color':'white','fontsize':20,'fontstyle':'italic'})
ax.set_xlabel('Value of x',fontdict={'color':'white','fontsize':20,'fontfamily':'serif'})
ax.set_ylabel('Value of y',fontdict={'color':'white','fontsize':20,'fontfamily':'serif'})
plt.show()
```

# We can draw multiple axes inside same figure.

```python
figure = plt.figure(figsize=(10,6),facecolor='blue')
x = np.arange(6)
y = x**2
z=x**3
ax1 = figure.add_axes([0.1,0.1,0.8,0.8]) #[l,b,w,h]
ax1.plot(x,y,'o--g',markersize=20)
ax1.set_title("x vs y",fontdict={'color':'white','fontsize':20,'fontstyle':'italic'})
ax1.set_xlabel('Value of x',fontdict={'color':'white','fontsize':20,'fontfamily':'serif'})
ax1.set_ylabel('Value of y',fontdict={'color':'white','fontsize':20,'fontfamily':'serif'})

ax1 = figure.add_axes([0.2,0.4,0.3,0.4]) #[l,b,w,h]
ax1.plot(x,z,'o--b',markersize=5)
plt.show()
```

## Subplot Function will be helpful.
- Easy to use and subplot() returns fig and axes.

```python
n_bins = 10
x = np.random.randn(1000, 3)
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2)
colors = ['red', 'tan', 'lime']
ax0.hist(x, n_bins, histtype='bar', color=colors, label=colors)
ax0.legend(prop={'size': 10})
ax0.set_title('bars with legend')
ax1.hist(x, n_bins, histtype='bar', stacked=True)
ax1.set_title('stacked bar')
ax2.hist(x, n_bins, histtype='step', stacked=True, fill=False)
ax2.set_title('stack step (unfilled)')
# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype='bar')
ax3.set_title('different sample sizes')
fig.tight_layout()
plt.show()
```
